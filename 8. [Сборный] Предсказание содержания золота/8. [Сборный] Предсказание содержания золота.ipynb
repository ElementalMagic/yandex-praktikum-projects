### Правильно ли рассчитана эффективность обогащения?

```python
def rougher_recovery(row):
    feed = row['rougher.input.feed_au']
    output_conc = row['rougher.output.concentrate_au']
    output_tails = row['rougher.output.tail_au']
    
    result = (output_conc * (feed - output_tails)) / (feed * (output_conc - output_tails)) * 100
    return result
```

```python
df_train['rougher.output.recovery_calc'] = df_train.apply(rougher_recovery, axis=1)
df_train = df_train[(df_train['rougher.output.recovery_calc'] >= 0) & (df_train['rougher.output.recovery_calc'] <= 100)].reset_index(drop=True)
```

```python
sklearn.metrics.mean_absolute_error(df_train['rougher.output.recovery'], df_train['rougher.output.recovery_calc'])
```

#### Вывод

Эффективность обогащения сырья рассчитана правильно - величина ошибка является вычислительной погрешностью.

```python
def material_sum(df, columns, label):
    
    """Суммирование количество материалов и вывод гистограммы"""
    
    for (column, title) in zip(columns, ['feed', 'rougher', 'final']):
        df[f'materials_sum_{title}'] = df[column].sum(axis=1)
        df[f'materials_sum_{title}'].hist(label=title, alpha=0.5, bins=100)
        plt.legend()
    plt.title(label)
    plt.xlabel('Количество материала')
    plt.show()
```

```python
def remove_outliners(columns, df_full):
    
    """Удаление выбросов"""
    
    for column in columns:
        IQR = df_full[column].quantile(q=0.75) - df_full[column].quantile(q=0.25)
        lower = df_full[column].quantile(q=0.25) - 1.5*IQR
        upper = df_full[column].quantile(q=0.75) + 1.5*IQR
        df_full = df_full[(df_full[column] >= lower) & ((df_full[column] <= upper))].reset_index(drop=True)
    return df_full
```

```python
columns = [
    ['rougher.input.feed_ag', 'rougher.input.feed_pb', 'rougher.input.feed_sol', 'rougher.input.feed_au'],
    ['rougher.output.concentrate_ag', 'rougher.output.concentrate_pb', 'rougher.output.concentrate_sol', 'rougher.output.concentrate_au'],
    ['final.output.concentrate_ag', 'final.output.concentrate_pb', 'final.output.concentrate_sol', 'final.output.concentrate_au']
]

material_sum(df_train, columns, 'Распределение количества метриалов на разных стадиях обработки')
material_sum(df_test, [columns[0]], 'Распределение количества метриалов перед обработкой')
```

### Функция для вычисления итоговой sMAPE.

```python
def smape_loss(true, preds):
    assert(len(true) == len(preds)), 'Выборки содержат разное количество элементов'
    
    N = len(true)
    res = 0
    for i in range(0, N):
        res += (abs(true[i] - preds[i])) / ((abs(true[i]) + abs(preds[i])) / 2)
    return res * 100 / N
```

```python
smape_score = make_scorer(smape_loss, greater_is_better=False)
```

```python
def final_smape(rougher, final):
    return smape_loss(*rougher) * 0.25 + 0.75 * smape_loss(*final)
```

#### Линейная регрессия

```python
def linear_regression(columns, target_column):
    features = df_train[columns]
    target = np.array(df_train[target_column])

    scaler = StandardScaler()
    scaler.fit(features)
    features = scaler.transform(features)
    
    # Исключительно для примера
    param_grid_lr = {'normalize': [False, True]}

    clf = GridSearchCV(estimator=LinearRegression(),
                       param_grid=param_grid_lr,
                       cv=5,
                       verbose=0,
                       n_jobs=-1,
                       scoring=smape_score)
    best_linear = clf.fit(features, target)

    print('Линейная модель:')
    print(f'sMAPE = {best_linear.best_score_*(-1) : .3f}%')
    
    return best_linear
```

#### Случаный лес

```python
def random_forest(columns, target_column):
    features = df_train[columns]
    target = np.array(df_train[target_column])

    features_train, features_valid, target_train, target_valid = train_test_split(
        features, target, test_size=0.25, shuffle=True, random_state=12345)
    
    scaler = StandardScaler()
    scaler.fit(features_train)
    features_train = scaler.transform(features_train)
    
    model = RandomForestRegressor(n_estimators=100,
                          criterion='mae',
                          max_depth=20,
                          min_samples_split=2,
                          n_jobs=-1,
                          random_state=12345).fit(features_train, target_train)
    
    features_valid = scaler.transform(features_valid)

    print('Rnadom Forest Regressor:')
    print(f'sMAPE = {smape_loss(target_valid, model.predict(features_valid)):.3f}%')
    
    return model, scaler
```

#### Дерево решений

```python
def decision_tree(columns, target_column):
    features = df_train[columns]
    target = np.array(df_train[target_column])

    scaler = StandardScaler()
    scaler.fit(features)
    features = scaler.transform(features)

    param_grid_lr = {
        'criterion': ['mae'],
        'max_depth': [10, 20],
        'min_samples_split': np.arange(2, 10, 4),
        'min_samples_leaf': np.arange(1, 5, 2),
        'random_state': [564683]
    }

    clf = GridSearchCV(estimator=DecisionTreeRegressor(),
                       param_grid=param_grid_lr,
                       cv=5,
                       verbose=0,
                       n_jobs=-1,
                       scoring=smape_score)
    best_tree = clf.fit(features, target)

    print('Дерево решений:')
    print(f'sMAPE = {best_tree.best_score_*(-1):.3f}%')
    
    return best_tree
```
