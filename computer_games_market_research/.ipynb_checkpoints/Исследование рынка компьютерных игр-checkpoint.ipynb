## Описание данных и задача

### Задача

Используя исторические данные о продажах компьютерных игр, оценки пользователей и экспертов, жанры и платформы, выявить закономерности, определяющие успешность игры 

### Описание данных

* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)
* Other_sales — продажи в других странах (миллионы проданных копий)
* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Предобработка данных

Заменим названия столбцов.

```python
data.columns = ['name', 'platform', 'year_of_release', 'genre', 'na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'critic_score', 'user_score', 'rating']
```

Заполним пустые значения у оценок с помощью *tbd*.

```python
data['critic_score'].fillna('tbd', inplace=True)
data['user_score'].fillna('tbd', inplace=True)
data['rating'].fillna('tbd', inplace=True)
```

Пропуски в столбце `year_of_release` заполним медианным значением года выпуска для платформы.

```python
(data['year_of_release']
 .fillna(data.groupby('platform')['year_of_release']
 .transform('median'), inplace=True))
data = data.astype({'year_of_release': int})
```

```python
data = data.dropna(subset=['name']).reset_index(drop=True)
```

Посчитаем общие продажи игр и запишем в столбец `total_sales`.

```python
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
```

```python
data.head(10)
```

### Вывод

* Были заполнены пропуски в стобцах, рейтингом и оценками пользователей и годом релиза;
* Были удалены две игры без названия;
* Посчитаны общие продажи по всем регионам в столбце `total_sales`.

### Продажи по платформам

```python
games_per_platform = data.query('year_of_release != 0').pivot_table(index=['platform'], values=['total_sales'], aggfunc=['sum'])
games_per_platform.columns = ['total_sales']
games_per_platform.query('total_sales >= 200').sort_values(by='total_sales', ascending=False)
```

Считаю, что стоит разделить стационарные и портативные платформы. Эти аудитории часто не пересекаются, в каждой может быть популярен свой жанр и конкретные игры.

```python
platform_per_year = data.query('year_of_release != 0').pivot_table(index=['year_of_release', 'platform'], values=['total_sales'], aggfunc=['count', 'sum', 'mean'])
platform_per_year.columns = ['games_count', 'total_sales', 'mean_sales']
platform_per_year = platform_per_year.reset_index()
platform_per_year.head()
```

Определим функцию для вывода совместной столбчатой диаграммы.

```python
def stacked_plot(platforms, data, years):
    stats = data.query('year_of_release != 0').query('platform in @platforms').pivot_table(index=['year_of_release', 'platform'], values=['total_sales'], aggfunc=['sum'])

    years = [x for x in range(years[0], years[1])]
    sales_over_platforms_per_year = pd.DataFrame()

    for platform in platforms:
        platform_stats = []
        for year in years:
            try:
                count = stats.loc[(year, platform), 'sum'].values[0]
            except:
                count = 0
            platform_stats.append(count)
        sales_over_platforms_per_year[platform] = platform_stats

    sales_over_platforms_per_year.index = years
    sales_over_platforms_per_year.plot.bar(stacked=True, figsize=(20,10))
    plt.title('Продажи игр на платформах {} с {} по {} года'.format(', '.join(platforms), years[0], years[-1]))
    plt.ylabel('Миллионов копий')
    plt.xlabel('Год')
```

```python
interested_platforms_stationary = ['PS2', 'X360', 'PS3', 'Wii', 'PS','PS4', 'PC', 'XB']
for platform in interested_platforms_stationary:
    platform_per_year.query('platform == @platform').plot.bar(x='year_of_release', y='total_sales', label=platform)
    plt.xlabel('Календарный год')
    plt.ylabel('Млн. копий')
    plt.title('Продажи на платформе {} по годам'.format(platform))
```

В среднем, жизненный цикл современной успешной стационарной платформы составляет 10 лет. К концу своего жизненного цикла, платформа теряет аудиторию, пропадает поддержка от разработчиков как самой консоли, так и игр. Продажи стремятся к нулю.

```python
stacked_plot(interested_platforms_stationary, data, (1994, 2017))
```

### Анализ корреляции отзывов и продаж

За основу возьмем данные с консоли PlayStation 2, как самой популярной консоли в истории.

```python
playstation_2 = data.query('platform == "PS2"')
metrics = ['critic_score', 'user_score']
for metric in metrics:
    stats = playstation_2[playstation_2[metric] != 'tbd'].astype({metric: float})
    stats.plot(y='total_sales', x=metric, kind='scatter', alpha=0.1)
    plt.xlabel('Оценки')
    plt.ylabel('Продажи, млн. шт.')
    plt.title('Диаграмма рассеяния для total_sales и {}'.format(metric))
    plt.show()
    
    print('Коэффициент корреляции Пирсона между total_sales и {} = {:.3f}\n\n'.format(metric, stats['total_sales'].corr(stats[metric])))
```

Видеоигровые критики оценивают самые разные аспекты игры - от плавности, точность анимаций и художественного нарратива, ценности и структуры повествования до качества реплик, озвучки персонажей,  сбалансированности игровых механик и технического состояния. Критики стараются делать объективные выводы. Игроки же руководствуются сугубо своими ощущениями и впечатлениями от игры. Зачастую неискушенный игрок низко оценивает сложную и глубокую игру из-за того, что он её просто не понял. 

Критик способен точнее определить качество конечного продукта и поэтому оценки критиков лучше коррелируют с итоговыми продажами конкретной игры. Нередко можно обнаружить, что совсем неизвестная игра имеет пользовательский рейтинг близкий к 10 и 30 оценок, что говорят о малых продажах.

В целом положительная корреляция между пользовательскими оценками и продажами игры есть, но не особенно сильная. Корреляция между оценками критиков и продажами игры корреляция куда сильнее, можно утверждать, что игры, которые высоко оценены критиками, вероятно получат большие продажи, чем игры, которые критики оценили низко.

*Сравним полученные выводы с двумя другими популярными консолями - Xbox 360 и Nintendo DS*

```python
consoles = ['X360', 'DS']
for console in consoles:
    print('Статистика для консоли {}'.format(console))
    console_data = data.query('platform == @console')
    metrics = ['critic_score', 'user_score']
    for metric in metrics:
        stats = console_data[console_data[metric] != 'tbd'].astype({metric: float})
        stats.plot(y='total_sales', x=metric, kind='scatter', alpha=0.1)
        plt.xlabel('Оценки')
        plt.ylabel('Продажи, млн. шт.')
        plt.title('Диаграмма рассеяния для total_sales и {}'.format(metric))
        plt.show()

        print('Коэффициент корреляции Пирсона между total_sales и {} = {:.3f}\n\n'.format(metric, stats['total_sales'].corr(stats[metric])))
```

### Сравнение игровых жанров

Составим функцию аналогичную той, что использовалась для построения диаграмм продаж по годам. Добавим возможность посмотреть относительную долю рынка каждого жанра на диаграмме.

```python
def genre_plot(data, years, use_percent=False, show_final_table=False):
    stats = data.pivot_table(
            index=['year_of_release', 'genre'],
            values=['total_sales'],
            aggfunc=['sum'])

    years = [x for x in range(years[0], years[1])]
    genres_per_year = pd.DataFrame()

    genres = data['genre'].unique()
    
    for genre in genres:
        year_stats = []
        for year in years:
            try:
                if use_percent:
                    count = stats.loc[(year, genre), 'sum'].values[0] / data.query('year_of_release == @year')['total_sales'].sum() * 100
                else:
                    count = stats.loc[(year, genre), 'sum'].values[0]
            except:
                count = 0
            year_stats.append(count)
        genres_per_year[genre] = year_stats

    genres_per_year.index = years
    genres_per_year.plot.bar(stacked=True, figsize=(20, 10))
    if use_percent:
        plt.ylabel('%')
        plt.title('Процент рынка жанров {} с {} по {} года'.format(
        ', '.join(genres), years[0], years[-1]))
    else:
        plt.ylabel('Млн. копий')
        plt.title('Суммарные продажи игр жанров {} с {} по {} года'.format(
        ', '.join(genres), years[0], years[-1]))
    plt.xlabel('Год')
   
    if show_final_table:
        display(genres_per_year)
```

Посмотрим на статистику за весь период наблюдений.

```python
temp_columns = ['games_count', 'mean_sales', 'total_sales']
titles = [
    'Количество игр', 'Средние продажи', 'Суммарные продажи'
]
y_labels = ['Штук', 'Млн. копий', 'Млн. копий']

genres_over_time = data.pivot_table(index='genre',
                                    values=['total_sales'],
                                    aggfunc=['count', 'mean', 'sum'])
genres_over_time.columns = temp_columns
for i in range(0, 3):
    genres_over_time.plot.bar(y=temp_columns[i])
    plt.title(titles[i])
    plt.xlabel('Игровой жанр')
    plt.ylabel(y_labels[i])
    plt.show()
```

Отрыв от второго по популярности жанра **Sports** значительный -  *31%* по суммарным продажам и *43%* по созданным играм. Замыкает тройку лидеров жанр **Shooter** (по количеству проданных копий).

## Проверка гипотез

Для проверки гипотез будет использоваться метод `scipy.stats.ttest_ind()`, так как задача состоит в сравнении средних выборок двух генеральных совокупностей. Для проверки первой гипотезы параметр `equal_var`будет задан как `True`, потому что у нас нет оснований считать, что дисперсии двух выборок разные.

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Составим гипотезу $H_0$: средние пользовательские рейтинги платформ Xbox One и PC одинаковые  
Гипотеза $H_1$: средние пользовательские рейтинги платформ Xbox One и PC отличаются

Уровень значимости положим равным 5%.

```python
temp = data.query('user_score != "tbd"').astype({'user_score': float})
data1 = temp.query('platform == "XOne"')
data2 = temp.query('platform == "PC"')
results = st.ttest_ind(data1['user_score'], data2['user_score'], equal_var=True)
print('p-значение:', results.pvalue)
```

P-значение оказалось меньшим 0.05, что даёт основание отвергнуть гипотезу $H_0$ и принять гипотезу $H_1$.  
Вероятно, пользовательские рейтинги на платформах различаются.

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Составим гипотезу $H_0$: средние пользовательские рейтинги жанров Action и Sports одинаковые.  
Гипотеза $H_1$: средние пользовательские рейтинги жанров Action и Sports разные.  

Уровень значимости положим равным 5%.

```python
temp = data.query('user_score != "tbd"').astype({'user_score': float})
data1 = temp.query('genre == "Sports"')
data2 = temp.query('genre == "Action"')
results = st.ttest_ind(data1['user_score'], data2['user_score'], equal_var=True)
print('p-значение:', results.pvalue)
```

P-значение равно 10%, что выше статистической значимости в 5%, можно сделать вывод о том, что пользовательские рейтинги жанров Action и Sports могут и отличаться, но с вероятностью в 10% такие различия можно получить случайным образом, поэтому нет оснований отвергать нулевую гипотезу.
